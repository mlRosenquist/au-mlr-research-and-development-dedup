# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Metrics Bar Plot

```python
# set width of bars
barWidth = 0.3

fig, axs = plt.subplots(2, 2, figsize=(11,8))
ax1 = axs[0,0]
ax2 = axs[0,1]
ax3 = axs[1,0]
ax4 = axs[1,1]

###### Accuracy
# Syntehtic, Pendigits, Shuttle, WaveForm, WBC
acc_bars1 = [0.73, 0.54, 0.91, 0.92, 0.94] # Original data
acc_bars2 = [0.85, 0.51, 0.91, 0.93, 0.93] # Bases
acc_bars3 = [0.9, 0.54, 0.87, 0.91, 0.93] # Bases extended

f1_bars1 = [0.25, 0.70, 0.95, 0.96, 0.97] # Original data
f1_bars2 = [0.23, 0.71, 0.95, 0.96, 0.97] # Bases
f1_bars3 = [0.40, 0.71, 0.94, 0.95, 0.97] # Bases extended

rec_bars1 = [1, 0.54, 0.93, 0.94, 0.94] # Original data
rec_bars2 = [1, 0.50, 0.92, 0.95, 0.93] # Bases
rec_bars3 = [1, 0.55, 0.9, 0.93, 00.93] # Bases extended

roc_bars1 = [0.86, 0.77, 0.46, 0.55, 0.97] # Original data
roc_bars2 = [0.92, 0.75, 0.46, 0.55, 0.97] # Bases
roc_bars3 = [0.95, 0.77, 0.45, 0.54, 0.97] # Bases extended

# Set position of bar on X axis
r1 = np.arange(len(acc_bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot
ax1.bar(r1, acc_bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='original')
ax1.bar(r2, acc_bars2, color='#557f2d', width=barWidth, edgecolor='white', label='bases')
ax1.bar(r3, acc_bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='DupRes')

ax2.bar(r1, f1_bars1, color='#7f6d5f', width=barWidth, edgecolor='white', )
ax2.bar(r2, f1_bars2, color='#557f2d', width=barWidth, edgecolor='white',)
ax2.bar(r3, f1_bars3, color='#2d7f5e', width=barWidth, edgecolor='white', )

ax3.bar(r1, rec_bars1, color='#7f6d5f', width=barWidth, edgecolor='white', )
ax3.bar(r2, rec_bars2, color='#557f2d', width=barWidth, edgecolor='white', )
ax3.bar(r3, rec_bars3, color='#2d7f5e', width=barWidth, edgecolor='white', )

ax4.bar(r1, roc_bars1, color='#7f6d5f', width=barWidth, edgecolor='white', )
ax4.bar(r2, roc_bars2, color='#557f2d', width=barWidth, edgecolor='white', )
ax4.bar(r3, roc_bars3, color='#2d7f5e', width=barWidth, edgecolor='white', )

# Add xticks on the middle of the group bars
ax1.set_ylabel('accuracy', fontweight='bold')
ax2.set_ylabel('f1', fontweight='bold')
ax3.set_ylabel('recall', fontweight='bold')
ax4.set_ylabel('roc_auc', fontweight='bold')


ax3.set_xlabel('dataset', fontweight='bold')
ax4.set_xlabel('dataset', fontweight='bold')

ax1.set_xticks([r + barWidth for r in range(len(acc_bars1))], ['Synthetic', 'Pendigits', 'Shuttle', 'WaveForm', 'WBC'])
ax2.set_xticks([r + barWidth for r in range(len(acc_bars1))], ['Synthetic', 'Pendigits', 'Shuttle', 'WaveForm', 'WBC'])
ax3.set_xticks([r + barWidth for r in range(len(acc_bars1))], ['Synthetic', 'Pendigits', 'Shuttle', 'WaveForm', 'WBC'])
ax4.set_xticks([r + barWidth for r in range(len(acc_bars1))], ['Synthetic', 'Pendigits', 'Shuttle', 'WaveForm', 'WBC'])

# Create legend & Show graphic
lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, 'center right')

fig.suptitle("Performance metrics of models (d=2)")
fig.show()
```

# Time bar plot

```python
# set width of bars
barWidth = 0.3

bars1 = [78.8, 80.2, 90.4] # Original data
bars2 = [51.4, 51.2, 59.7] # Bases

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='training time')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='testing time')


# Add xticks on the middle of the group bars
plt.ylabel('time [ms]', fontweight='bold')

plt.xlabel('model', fontweight='bold')

plt.xticks([r + barWidth for r in range(len(bars1))], ['original', 'bases', 'DupRes'])

# Create legend & Show graphic
plt.legend(['Training time', 'Testing time'])

plt.title("Execution time of models (d=1)")
fig.show()
```
